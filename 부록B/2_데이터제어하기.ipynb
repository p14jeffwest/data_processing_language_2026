### 2.1. 조건문 (if, elif, else)

```python
score = 85

if score >= 90:
    print("A학점")
elif score >= 80:
    print("B학점")
else:
    print("C학점")
```

### 2.2. 반복문 (for, while) - 데이터 순회하기 [핵심]

`for`는 리스트에 있는 데이터를 하나씩 꺼내서 처리할 때 가장 많이 쓴다.

```python
# 1. 리스트 순회
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(f"I like {fruit}")
```

```python
# 2. 횟수 반복 (range)
# 0부터 4까지 5번 반복
for i in range(5):
    print(i) # 0, 1, 2, 3, 4
```

`while`은 특정 조건이 만족할 때까지 수행하는 경우 사용한다. 

```python
# 예시: 1부터 5까지 출력
i = 1
while i <= 5:
    print(i)
    i += 1
```

`for`와 `while` 루프내에서 `break`와 `continue`를 사용할 수 있다. 

```python
num = 0

while True:
    if num >= 10:
        break
    num += 1
    if num % 2 == 0:
        continue
    print(num)
```

### 2.3. 똑똑한 반복문 (enumerate, zip) [실무 필수]

데이터 분석에서 두 개 이상의 리스트를 다루거나, 순서(인덱스)가 필요할 때 사용하는 필수 기법이다.

```python
# 1. enumerate: 순서(인덱스)와 데이터를 같이 꺼내기
# "몇 번째 데이터인지" 알고 싶을 때 씁니다.
names = ["철수", "영희", "민수"]

for i, name in enumerate(names):
    print(f"{i+1}번 학생: {name}")
# 출력: 1번 학생: 철수, 2번 학생: 영희...
```

```python
# 2. zip: 두 개의 리스트를 지퍼처럼 묶기
# 날짜 리스트와 가격 리스트를 동시에 처리할 때 씁니다.
dates = ["1월 1일", "1월 2일"]
prices = [1000, 2000]

for date, price in zip(dates, prices):
    print(f"{date}의 가격은 {price}원")
# 출력: 1월 1일의 가격은 1000원...
```

## TIP : 리스트 내포 (List Comprehension)

데이터 전처리 시 코드를 한 줄로 줄여주는 파이썬만의 강력한 기능입니다. 처음엔 어색할 수 있지만, 익숙해지면 코드가 훨씬 간결해집니다.

```python
# 일반적인 방식
nums = [1, 2, 3, 4, 5]
squares = []
for n in nums:
    squares.append(n * 2)

print(squares) # [2, 4, 6, 8, 10]
```

```python
# 리스트 내포 방식 (위와 똑같은 결과)
squares = [n * 2 for n in nums] 
# 해석: nums에 있는 n을 하나씩 꺼내서 n*2를 한 뒤 리스트로 만듦

print(squares) # [2, 4, 6, 8, 10]
```
