### 1.1. 변수와 기본 자료형

파이썬은 변수 선언 시 타입을 지정하지 않아도 된다(`int a` 같은 선언 불필요).

```python
# 1. 숫자형 (Integer, Float)
age = 25
height = 175.5
print(age + 5)  # 덧셈, 뺄셈, 곱셈(*), 나눗셈(/) 가능
```

```python
# 2. 문자열 (String) - 따옴표('', "")로 감싼다
name = "홍길동"
greeting = 'Hello'
print(name + "님, " + greeting)  # 문자열 연결 가능
```

```python
# 3. 불리언 (Boolean) - 참/거짓
is_student = True
is_adult = False

print("학생인가?", is_student)
print("성인인가?", is_adult)
```

```python
# f-string을 사용한 문자열 포매팅
name = "홍길동"
age = 25
print(f"제 이름은 {name}이고, 나이는 {age}세입니다.")
```

### 1.2. 리스트 (List) - 데이터의 줄 세우기 [중요]

엑셀의 '한 열(Column)'과 같다. 순서가 있으며 수정 가능하다. 대괄호 `[]`를 쓴다.

```python
# 리스트 생성
colors = ["red", "green", "blue"]

# 인덱싱 (Indexing): 0부터 시작한다!
print(colors[0])  # red
print(colors[-1]) # blue (뒤에서 첫 번째)
```

```python
# 슬라이싱 (Slicing): [시작:끝] (끝 번호는 포함 안 됨)
print(colors[0:2]) # ['red', 'green'] (0번, 1번만 나옴)
```

```python
# 추가와 삭제
colors.append("yellow") # 맨 뒤에 추가
colors.remove("green")  # 삭제

print(colors) # ['red', 'blue', 'yellow']
```

### 1.3. 튜플 (Tuple) - 변하지 않는 리스트

리스트와 거의 같지만, **한번 생성하면 내용을 바꿀 수 없다(Immutable)**는 점이 다르다. **소괄호 `()`**를 쓴다. 데이터가 실수로 변경되는 것을 막고 싶을 때 사용한다.

```python
# 튜플 생성 (좌표, 설정값 등에 사용)
coords = (37.5, 127.0)

# 인덱싱 (가능)
print(coords[0]) # 37.5
```

```python
# 수정 (불가능 - 에러 발생!)
coords[0] = 38.0  # TypeError 발생
```

### 1.4. 집합 (Set) - 중복 없는 주머니

순서가 없고, **중복된 데이터가 허용되지 않는** 자료형이다. **중괄호 `{}`**를 쓴다. 주로 리스트에서 중복을 제거하거나, 교집합/합집합을 구할 때 쓴다.

```python
# 집합 생성 (중복은 자동으로 제거됨)
nums = {1, 2, 3, 3, 3}
print(nums) # {1, 2, 3}
```

```python
# 교집합(&)과 합집합(|)
set_a = {1, 2, 3}
set_b = {2, 3, 4}

print(set_a & set_b) # {2, 3} (교집합)
print(set_a | set_b) # {1, 2, 3, 4} (합집합)
```

### 1.5. 딕셔너리 (Dictionary) - 이름표 붙이기 [중요]

JSON 데이터나 엑셀의 '한 행(Row)'을 표현할 때 쓴다. **중괄호 `{}`**를 쓰며 `Key: Value` 쌍으로 이루어진다.

```python
student = {
    "name": "홍길동",
    "id": 2026001,
    "grade": 4.5
}

# 데이터 꺼내기 (Key를 이용)
print(student["name"]) # 홍길동
```

```python
# 데이터 추가/수정
student["major"] = "산업공학" # 새로운 키 추가
student["grade"] = 4.0        # 기존 값 수정
```
