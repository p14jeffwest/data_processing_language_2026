### 3.1. 함수 (Function) 
복잡한 로직을 하나의 명령어처럼 묶는다. 함수는 **입력(매개변수)을 받아 처리 후 출력(반환값)을 내놓는 자판기**와 같다.

#### (1) 기본 구조와 반환값 (return)

파이썬에서의 함수 기본 구조는 아래와 같다. 

```python
def 함수명(매개변수1, 매개변수2, ...):
    <수행할 문장1>
    <수행할 무장2>
    ...
    return 반환값 # return은 생략 가능 (None 반환)
```

`return`은 함수가 계산한 최종 결과를 호출한 곳으로 '던져주는' 역할을 한다.

````python
def calculate_area(width, height):
    result = width * height
    return result  # 계산된 값을 밖으로 반환

# 사용하기: 반환값을 변수(area)에 저장할 수 있다
area = calculate_area(10, 20)
print(f"면적은 {area}이다.")
````

#### (2) 디폴트 매개변수 (Default Parameter)

매개변수에 기본값(Default)을 미리 정해줄 수 있다. 값을 넣지 않으면 기본값이 사용되고, 값을 넣으면 새 값으로 덮어쓴다.

````python
# message의 기본값을 "안녕하세요"로 설정
def hello(name, message="안녕하세요"):
    print(f"{name}님, {message}")

# 1. 기본값 사용 (이름만 입력)
hello("홍길동") 
# 출력: 홍길동님, 안녕하세요
````

````python
# 2. 값 변경 (메시지도 입력)
hello("홍길동", "반갑습니다") 
# 출력: 홍길동님, 반갑습니다
````

### 3.2. 모듈(라이브러리) 사용하기 [필수]
파이썬이 강력한 이유는 남들이 만들어 둔 좋은 도구(라이브러리)가 많기 때문이다. 

#### 파이썬 실행환경에 외부 모듈 추가하기

파이썬에 있는 기본 모듈은 그대로 `import`를 이용해서 사용하면되지만 외부 모듈의 경우는 파이썬 실행환경에 추가해줘야 한다. 

* **uv add <외부 모듈명>**  : `uv`를 사용하는 경우는 프로젝트 폴더에서 이와 같은 명령으로 추가할 수 있다.
* **pip install <외부 모듈명>** : `uv` 없이 기본 파이썬만 사용하는 경우는 `pip install` 명령을 사용한다. 

#### 모듈 사용하기

`import <모듈명>`를 이용해서 사용할 수 있다. 

````python
# 1. 수학 모듈 불러오기 (math는 파이썬의 기본 모듈이다. uv add 등을 통해 추가할 필요 없다.
import math
print(math.sqrt(16)) # 4.0
````

````python
# 2. 외부 모듈 불러오기 (pandas, numpy는 외부 모듈이다. 사용전에 uv add를 통해 추가해야한다. 
# 아래 예에서는 별명(Alias) 붙여서 불러오기했다. (일반적으로 이렇게 단축된 별명을 사용한다.)
import pandas as pd
import numpy as np
````

Terminal에서 `uv add numpy`와 `uv add pandas`를 한다. 

````python
import pandas as pd
import numpy as np
````

### 3.3. 예외 처리 (try-except) 

데이터 수집 중 에러가 나도 프로그램이 멈추지 않게 한다.

````python
text = "100달러"

try:
    # 숫자로 바꿀 수 없는 문자열을 숫자로 바꾸려 시도
    number = int(text) 
    print(number)
except ValueError:
    print("에러 발생: 숫자가 아닌 문자가 포함되어 있다.")

print("프로그램이 죽지 않고 계속 실행된다.")
````
